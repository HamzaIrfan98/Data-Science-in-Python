```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gradient_descent(x,y):
    curr_slope = curr_intercept = 0
    rate = 0.01
    n = len(x)
    plt.scatter(x,y, color='red', marker='+', linewidth=5)

    for i in range(10000):
        y_pred = curr_slope * x + curr_intercept
        plt.plot(x, y_pred, color='blue')
        md = -(2/n)*sum(x*(y-y_pred))
        yd = -(2/n)*sum(y-y_pred)
        curr_slope = curr_slope - rate * md
        curr_intercept = curr_intercept - rate * yd
```

```python
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([4,7,10,12,15,18,20,22])
```

```python
gradient_descent(x,y)
```
