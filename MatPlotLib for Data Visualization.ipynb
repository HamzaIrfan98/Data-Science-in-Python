```python
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline    # Will allow us to see our graph here in the jupyter notebook
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
```

```python
y=x**2
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

# 2nd Method (It is object oriented)

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')
axes1.set_title('Axes 1')

axes2.plot(y,x,'b')
axes2.set_xlabel('X axis')
axes2.set_ylabel('Y axis')
axes2.set_title('Axes 2')
```

```python
fig,axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Axes 1')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
```

# Legends, Labels, and Title

```python
ax.set_title('Title')

ax.set_xlabel('x')
ax.set_ylabel('y')

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label='x**2')
ax.plot(x, x**3, label='x**3')
ax.legend()
```

```python
fig,ax = plt.subplots()

ax.plot(x, x+1, color='blue', alpha=0.5)
ax.plot(x, x+2, color='#8B008B')
ax.plot(x, x+3, color='#FFBC00')
```

# Line & Marker Style

```python
fig, ax =plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color='red', linewidth=0.25)
ax.plot(x, x+2, color='red', linewidth=0.5)
ax.plot(x, x+3, color='red', linewidth=1)
ax.plot(x, x+4, color='red', linewidth=2)

ax.plot(x, x+5, color='green', lw=3, ls='-')
ax.plot(x, x+6, color='green', lw=3, ls='-.')
ax.plot(x, x+7, color='green', lw=3, ls=':')


line, = ax.plot(x, x+8, color='black', lw=1.5)
line.set_dashes([5,10,15,10])

ax.plot(x,x+9, color='blue', lw=3, ls='-', marker='o', markersize=10, markerfacecolor='red', markeredgewidth=3, markeredgecolor='green')
```

# Axis Appeareances

```python
fig, axes =plt.subplots(1,3, figsize=(12,4))

axes[0].plot(x,x**2,x,x**3)
axes[0].set_title('Axes[0]')

axes[1].plot(x,x**2,x,x**3)
axes[1].axis('tight')
axes[1].set_title('Tight axes: Axes[1]')

axes[2].plot(x,x**2,x,x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([2,5])
axes[2].set_title('Axes[2]')
```
